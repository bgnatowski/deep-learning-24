```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from pathlib import Path

np.random.seed(42) 

# Domyślne rozmiary czcionek służące do generowania ładnych rysunków
plt.rc('font', size=12)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=12)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize = 10)

# Ustawienia biblioteki Pandas
pd.set_option('display.max_columns', None) # wszystkie kolumny w dataframe
```

```python
datapath = Path() / "data" / "lifesat" # Definiujemy ścieżkę do bieżącego katalogu
# Wczytujemy zbiory
oecd_bli = pd.read_csv(datapath / "oecd_bli.csv")
gdp_per_capita = pd.read_csv(datapath / "gdp_per_capita.csv")
```

```python
oecd_bli
```

```python
gdp_per_capita
```

```python

# Wstepnie przetwarzamy gdp za rok 2020
gdp_per_capita_2020 = gdp_per_capita[gdp_per_capita["Year"]==2020]
gdp_per_capita_2020 = gdp_per_capita_2020.rename(columns={"GDP per capita, PPP (constant 2017 international $)" : "GDP (USD)"})
gdp_per_capita_2020
```

```python
# Poszukiwanie przez nas zadowolenie z życia (Life satisfaction) to jedn
oecd_bli["Indicator"].value_counts()
```

```python
# Dlatego musimy utworzyc tablele przestawną (pivot table) z interesujacymi nas wartościami
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"] == "TOT"]
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
oecd_bli
```

```python
# Zresetujemy indeks, aby kraj był z powrotem kolumną. Przyda nam sie to do łączenia tabel
oecd_bli = oecd_bli.reset_index()
oecd_bli
```

```python
# Wyswielamy podstawowe informacje o dataframe
gdp_per_capita_2020.info()
```

```python
# Wyswielamy podstawowe informacje o dataframe
oecd_bli.info()
```

```python
# kolumny Country i Entity sa tego samego typu - robimy merge'a
full_country_stats = oecd_bli.merge(gdp_per_capita_2020, left_on="Country", right_on="Entity", how="inner")
full_country_stats
```

```python
# Ograniczamy sie do interesujacych nas kolumn i sortujemy
full_country_stats = full_country_stats[["Country", "Life satisfaction", "GDP (USD)"]]
full_country_stats = full_country_stats.sort_values(by="GDP (USD)", ascending=False)
full_country_stats
```

```python
full_country_stats.to_csv(datapath / "lifestat_full.csv", index=False)
```

```python
min_gdp = 23_500
max_gdp = 62_500

country_stats = full_country_stats[(full_country_stats["GDP (USD)"]>=min_gdp) & (full_country_stats["GDP (USD)"]<=max_gdp)]
country_stats.to_csv(datapath / "lifestat.csv", index=False)
```

```python
lifestat = pd.read_csv(datapath / "lifestat.csv")
lifestat
```

```python
# wizualizujemy dane
lifestat.plot(kind='scatter', grid=True, x="GDP (USD)", y="Life satisfaction")
plt.show()
```

```python
# Przygotowanie zmiennych w modelu / mozna uzyc .to_numpy()
X = lifestat[["GDP (USD)"]]
y = lifestat[["Life satisfaction"]]

display(type(X))
display(type(y))

# Wybiera model liniowy
model = LinearRegression()

# Trenuje model
model.fit(X, y)

# Obliczone współczynniki modeli
theta_0 = model.intercept_[0]
theta_1 = model.coef_[0][0]
print(f"theta_0: {theta_0}")
print(f"theta_1: {theta_1}")
```

```python
# wizualiacja danych
lifestat.plot(kind="scatter", grid=True, x="GDP (USD)", y="Life satisfaction")
plt.plot(X, theta_0+theta_1*X)
plt.text(45000, 6.5, fr"$\theta_0 = {theta_0.round(2)}", color="blue", fontweight="bold")
plt.text(45000, 6.5, fr"$\theta_1 = {theta_1}", color="blue", fontweight="bold")
plt.show()
```

```python
cyprus_gdp_per_capita_2020 = (
    gdp_per_capita_2020[gdp_per_capita_2020["Entity"]=="Cyprus"][["GDP (USD)"]]
)
cyprus_gdp_per_capita_2020
```

```python
cyprus_predicted_life_satisfaction = model.predict(cyprus_gdp_per_capita_2020)
cyprus_predicted_life_satisfaction
```
